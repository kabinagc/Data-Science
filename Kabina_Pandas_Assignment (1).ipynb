### 2.

The dataset we'll be working with has the following columns:

|column      |description |
|:-------------|:-----------|
|country       | Country Name |
|food_category | Food Category |
|consumption   | Consumption (kg/person/year) |
|co2_emmission | Co2 Emission (Kg CO2/person/year) |


Import the dataset as a dataframe named `df` from this url: <https://raw.githubusercontent.com/rfordatascience/tidytuesday/master/data/2020/2020-02-18/food_consumption.csv>

```python
# Your answer here.
df = pd.read_csv('https://raw.githubusercontent.com/rfordatascience/tidytuesday/master/data/2020/2020-02-18/food_consumption.csv')
df.head()
```

### 3.

How many rows and columns are there in the dataframe?

```python
# Your answer here.
columns=df.columns
rows=df.shape[0]
print(columns, rows)
```

### 4.

What is the type of data in each column of `df`?

```python
# Your answer here.
col_types=df.dtypes
print(col_types)
```

### 5.

What is the mean `co2_emission` of the whole dataset?

```python
# Your answer here.
mean = df['co2_emmission'].mean()
print(mean)
```

### 6.

How many different kinds of foods are there in the dataset? How many countries are in the dataset?

```python
# Your answer here.

uni_food = df['food_category'].nunique()

uni_countries = df['country'].nunique()
print(uni_food,uni_countries)
```

### 7.

What is the maximum `co2_emmission` in the dataset and which food type and country does it belong to?

```python
# Your answer here.
max_co2_row = df.loc[df['co2_emmission'].idxmax()]

max_co2_emmission = max_co2_row['co2_emmission']
food_type = max_co2_row['food_category']
country = max_co2_row['country']

# Print the results
print(max_co2_emmission,food_type, country )
```

### 8.

How many countries produce more than 1000 Kg CO2/person/year for at least one food type?

```python
# Your answer here.

high_co2_countries = df[df['co2_emmission'] > 1000]

num_high_co2_countries = high_co2_countries['country'].nunique()

print(num_high_co2_countries)
```

### 9.

Which country consumes the least amount of beef per person per year?

```python
# Your answer here.

beef_data = df[df['food_category'] == 'Beef']

min_beef_country = beef_data.loc[beef_data['co2_emmission'].idxmin()]['country']

print(min_beef_country)
```

### 10.

Which country consumes the most amount of soybeans per person per year?

```python
# Your answer here.

soybeans_data = df[df['food_category'] == 'Soybeans']

max_soybeans_country = soybeans_data.loc[soybeans_data['co2_emmission'].idxmax()]['country']

print(max_soybeans_country)
```

### 11.

What is the total emissions of all the meat products (Pork, Poultry, Fish, Lamb & Goat, Beef) in the dataset combined?

```python
# Your answer here.

meat_products = ['Pork', 'Poultry', 'Fish', 'Lamb & Goat', 'Beef']

meat_data = df[df['food_category'].isin(meat_products)]

total_meat_emissions = meat_data['co2_emmission'].sum()

print(total_meat_emissions)
```

### 12.

What is the total emissions of all other (non-meat) products in the dataset combined?

```python
# Your answer here.

meat_products = ['Pork', 'Poultry', 'Fish', 'Lamb & Goat', 'Beef']

non_meat_data = df[~df['food_category'].isin(meat_products)]

total_non_meat_emissions = non_meat_data['co2_emmission'].sum()

print(total_non_meat_emissions)
```
